## Step 3: Data Cleaning
Checking and handling missing values in the dataset.

Missing values found:
- director: 2634 missing
- cast: 825 missing
- country: 831 missing
- date_added: 10 missing
- rating: 4 missing
- duration: 3 missing

Strategy:
- director, cast, country, rating → filled with "Unknown" / "Not Rated"
- date_added, duration → dropped rows as these are important for analysis

```python
# Check missing values
print(df.isnull().sum())

# Fix missing values
df['director'] = df['director'].fillna('Unknown')
df['cast'] = df['cast'].fillna('Unknown')
df['country'] = df['country'].fillna('Unknown')
df['rating'] = df['rating'].fillna('Not Rated')

# Drop rows with missing date and duration
df.dropna(subset=['date_added', 'duration'], inplace=True)

# Confirm
print(df.isnull().sum())
```

## Step 6: SQL Practice
In this step we will run SQL queries on our Netflix data
using SQLite — a lightweight database built into Python
No installation needed!

```python
import sqlite3

# Create database in memory
conn = sqlite3.connect(':memory:')

# Load dataframe into SQL table
df.to_sql('netflix', conn, index=False)

print("Data loaded into SQL successfully!")
```

```python
# Reload dataframe into SQL with year_added column
df.to_sql('netflix', conn, index=False, if_exists='replace')

print("Data reloaded successfully!")
```

### SQL Query 1: Movies vs TV Shows

```python
query1 = pd.read_sql_query(
    """
    SELECT type, COUNT(*) AS total_count
    FROM netflix
    GROUP BY type
    """,
    conn
)
print(query1
)
```

### SQL Query 2: Top 10 Countries by Content

```python
query2 = pd.read_sql_query(
    """
    SELECT country, COUNT(*) AS total_count
    FROM netflix
    WHERE country != 'Unknown'
    GROUP BY country
    ORDER BY total_count DESC
    LIMIT 10
    """,
    conn
)

print(query2)
```

### SQL Query 3: Netflix Content Added Per Year

```python
query3 = pd.read_sql_query(
    """
    SELECT year_added, COUNT(*) AS total_count
    FROM netflix
    WHERE year_added IS NOT NULL
    GROUP BY year_added
    ORDER BY year_added
    """,
    conn
)

print(query3)
```

### SQL Query 4: Top 10 Countries by Content

```python
query4 = pd.read_sql_query(
    """
    SELECT country, COUNT(*)  AS total_count
    FROM netflix
    WHERE country != 'Unknown'
    GROUP BY country
    ORDER BY total_count DESC
    """,
    conn
)
print(query4)
```

### SQL Query 5: Most Prolific Directors on Netflix

```python
query5 = pd.read_sql_query(
    """
    SELECT Director, COUNT(*) AS total_count
    FROM netflix
    WHERE director != 'Unknown'
    Group by Director
    Order by total_count DESC
    """,
    conn
)

print(query5)
```
